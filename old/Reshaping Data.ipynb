```python
import pandas as pd
from functools import reduce
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salary = pd.read_csv("merged_by_year_salary.csv")
salary.head()
```

```python
salary_age = salary.iloc[:, :11]
salary_age.head()
```

```python
column_indices_race = [0] + list(range(11, 15))
salary_race = salary.iloc[:, column_indices_race]
salary_race.head()
```

```python
column_indices_education = [0] + list(range(15, 19))
salary_education = salary.iloc[:, column_indices_education]
salary_education.head()
```

```python
## function to transform generic dataset
def transform(df):
    num_entries = len(df) * len(df.columns[1:])
    cols = {'Year':[],'Percentage': []}
    for i in df.columns[1:]:
        cols[i] = [0] * num_entries
    for i in range(len(df["Year"])):
        for j in range(1, len(df.columns)):
            cols['Year'].append(df['Year'][i])
            cols['Percentage'].append(df.iloc[i, j])
            cols[df.columns[j]][4*i+j-1] = 1
    return pd.DataFrame.from_dict(cols)
```

```python
salary_age_transformed = transform(salary_age)
salary_age_transformed.head()
```

```python
salary_age_transformed.columns
```

```python
salary_age_transformed = salary_age_transformed[['Year', 'Percentage','16 to 19 years', '20 to 24 years', '25 to 34 years', '35 to 44 years',
       '45 to 54 years', '55 to 64 years', '65 years and older',
       'Total 16 to 24', 'Total 25 and older', 'Total,\n16 years\nand older',]]
salary_age_transformed.head()
```

```python
salary_age_transformed.to_csv("salary_age.csv", index=False)
```

```python
salary_race_transformed = transform(salary_race)
salary_race_transformed.head()
```

```python
salary_race_transformed.columns
```

```python
salary_race_transformed = salary_race_transformed[['Year', 'Percentage','Asian', 'Black or African American', 'Hispanic or Latino ethnicity','White']]
salary_race_transformed.head()
```

```python
salary_race_transformed.to_csv("salary_race.csv", index=False)
```

```python
salary_education_transformed = transform(salary_education)
salary_education_transformed.head()
```

```python
salary_education_transformed.columns
```

```python
salary_education_transformed = salary_education_transformed[['Year', 'Percentage','Bachelor\'s\ndegree and\nhigher', 'High school\ngraduates, no\ncollege',
       'Less than a\nhigh school\ndiploma','Some college or associate degree']]
salary_education_transformed.head()
```

```python
salary_education_transformed.to_csv("salary_education.csv", index=False)
```

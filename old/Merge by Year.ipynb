```python
import pandas as pd
from functools import reduce
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table11 = pd.read_csv("below_min_wage_table11.csv")
table12 = pd.read_csv("salary_women_percentage_of_men_table12.csv")
table13 = pd.read_csv("salary_by_race_table13.csv")
table14 = pd.read_csv("salary_by_education_table14.csv")
table15 = pd.read_csv("hourly_women_percentage_of_men_table15.csv")
table16 = pd.read_csv("hourly_by_race_table16.csv")

merge_dfs = [table11, table12.drop("Year", axis=1), table13.drop("Year", axis=1), table14.drop("Year", axis=1), table15.drop("Year", axis=1), table16.drop("Year", axis=1)]
merge_dfs_salary = [table12, table13.drop("Year", axis=1), table14.drop("Year", axis=1)]
merge_dfs_hourly = [table15, table16.drop("Year", axis=1)]
```

```python
display(table11.head())
display(table12.head())
display(table13.head())
display(table14.head())
display(table15.head())
display(table16.head())
```

```python
df_merged = pd.concat(merge_dfs, join='inner', axis=1)
df_merged
```

```python
df_merged.to_csv("merged_by_year.csv", index=False)
```

```python
df_merged_salary = pd.concat(merge_dfs_salary, join='inner', axis=1)
df_merged_salary["Asian"] = df_merged_salary["Asian"].replace("–", 79.9) #backfill with earliest value pre 2000
df_merged_salary
```

```python
df_merged_salary.columns
```

```python
df_merged_hourly = pd.concat(merge_dfs_hourly, join='inner', axis=1)
df_merged_hourly['Asian'] = df_merged_hourly['Asian'].replace("–", 90.5) #backfill with earliest value pre 2000
df_merged_hourly
```

```python
df_merged_salary.to_csv("merged_by_year_salary.csv", index=False)
df_merged_hourly.to_csv("merged_by_year_hourly.csv", index=False)
```
